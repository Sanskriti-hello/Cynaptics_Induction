```python
#Appropriate Modules Imported
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import numpy as np
import time
#Making The Generator Model
def make_generator_model():
    return nn.Sequential(
        nn.Linear(100, 7 * 7 * 256),
        nn.BatchNorm1d(7 * 7 * 256),
        nn.LeakyReLU(0.2),

        nn.Unflatten(1, (256, 7, 7)),
        nn.ConvTranspose2d(256, 128, kernel_size=5, stride=1, padding=2, bias=False),
        nn.BatchNorm2d(128),
        nn.LeakyReLU(0.2),

        nn.ConvTranspose2d(128, 64, kernel_size=5, stride=2, padding=2, output_padding=1, bias=False),
        nn.BatchNorm2d(64),
        nn.LeakyReLU(0.2),

        nn.ConvTranspose2d(64, 1, kernel_size=5, stride=2, padding=2, output_padding=1, bias=False),
        nn.Tanh()
    )
#Making the Discriminator Model
def make_discriminator_model():
    return nn.Sequential(
        nn.Conv2d(1, 64, kernel_size=5, stride=2, padding=2),
        nn.LeakyReLU(0.2),
        nn.Dropout(0.3),

        nn.Conv2d(64, 128, kernel_size=5, stride=2, padding=2),
        nn.LeakyReLU(0.2),
        nn.Dropout(0.3),

        nn.Flatten(),
        nn.Linear(128 * 7 * 7, 1)
    )
#It should be creating more images on the basis of getting info from noise
def gen_noise(batch_size, noise_dim, device):
    return torch.randn(batch_size, noise_dim, device=device)
#It should be training the generator according to the number of images
def train_step(real_images, generator, discriminator, gen_opt, disc_opt, loss_func, device):
    batch_size = real_images.size(0)
    noise_dim = 100

#Train The Discriminator
    disc_opt.zero_grad()
    noise = gen_noise(batch_size, noise_dim, device)
    fake_images = generator(noise)

    real_preds = discriminator(real_images)
    fake_preds = discriminator(fake_images.detach())

    real_loss = loss_func(real_preds, torch.ones_like(real_preds))
    fake_loss = loss_func(fake_preds, torch.zeros_like(fake_preds))
    disc_loss = (real_loss + fake_loss) / 2

    disc_loss.backward()
    disc_opt.step()

#Train The Generator
    gen_opt.zero_grad()
    noise = gen_noise(batch_size, noise_dim, device)
    fake_images = generator(noise)
    fake_preds = discriminator(fake_images)

    gen_loss = loss_func(fake_preds, torch.ones_like(fake_preds))

    gen_loss.backward()
    gen_opt.step()

    return gen_loss.item(), disc_loss.item()

def main():
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

#Load MNIST dataset
    transform = transforms.Compose([
        transforms.ToTensor(),
        transforms.Normalize((0.5,), (0.5,))
    ])

    train_dataset = datasets.MNIST(root="./data", train=True, transform=transform, download=True)
    train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)


    generator = make_generator_model().to(device)
    discriminator = make_discriminator_model().to(device)

    gen_opt = optim.Adam(generator.parameters(), lr=1e-4)
    disc_opt = optim.Adam(discriminator.parameters(), lr=1e-4)


    loss_func = nn.BCEWithLogitsLoss()

# Training The Parameters
    noise_dim = 100
    num_examples_to_generate = 16
    seed = gen_noise(num_examples_to_generate, noise_dim, device)

# Training The Loop
    start_time = time.time()
    for epoch in range(EPOCHS):
        epoch_start_time = time.time()
        gen_loss_avg = 0
        disc_loss_avg = 0
        num_batches = 0

        for real_images, _ in train_loader:
            real_images = real_images.to(device)
            gen_loss, disc_loss = train_step(real_images, generator, discriminator, gen_opt, disc_opt, loss_func, device)

            gen_loss_avg += gen_loss
            disc_loss_avg += disc_loss
            num_batches += 1

        gen_loss_avg /= num_batches
        disc_loss_avg /= num_batches

        epoch_elapsed_time = time.time() - epoch_start_time
        print(f"Epoch {epoch+1}/{EPOCHS} - Generator Loss: {gen_loss_avg:.4f}, Discriminator Loss: {disc_loss_avg:.4f}, Time: {epoch_elapsed_time:.2f}s")

#Generate and save images after every 10 epochs
        if (epoch + 1) % 10 == 0:
            with torch.no_grad():
                generated_images = generator(seed).cpu()
                generated_images = (generated_images * 0.5 + 0.5).clamp(0, 1).numpy()

            fig, axes = plt.subplots(4, 4, figsize=(4, 4))
            for i, ax in enumerate(axes.flat):
                ax.imshow(generated_images[i, 0], cmap="gray")
                ax.axis("off")
            plt.savefig(f"generated_images_epoch_{epoch+1}.png")
            plt.show()

    total_elapsed_time = time.time() - start_time
    print(f"Training completed in {total_elapsed_time:.2f}s")
#Calling the Main Function
if __name__ == "__main__":
    BATCH_SIZE = 128
    EPOCHS = 50
    main()
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import numpy as np
import time
from PIL import Image

def make_generator_model():
    return nn.Sequential(
        nn.Linear(100, 8 * 8 * 256),
        nn.BatchNorm1d(8 * 8 * 256),
        nn.LeakyReLU(0.2),

        nn.Unflatten(1, (256, 8, 8)),
        nn.ConvTranspose2d(256, 128, kernel_size=4, stride=2, padding=1, bias=False),
        nn.BatchNorm2d(128),
        nn.LeakyReLU(0.2),

        nn.ConvTranspose2d(128, 64, kernel_size=4, stride=2, padding=1, bias=False),
        nn.BatchNorm2d(64),
        nn.LeakyReLU(0.2),

        nn.ConvTranspose2d(64, 3, kernel_size=4, stride=2, padding=1, bias=False),
        nn.Tanh()
    )

def make_discriminator_model():
    return nn.Sequential(
        nn.Conv2d(3, 64, kernel_size=4, stride=2, padding=1),
        nn.LeakyReLU(0.2),
        nn.Dropout(0.3),

        nn.Conv2d(64, 128, kernel_size=4, stride=2, padding=1),
        nn.LeakyReLU(0.2),
        nn.Dropout(0.3),

        nn.Flatten(),
        nn.Linear(128 * 16 * 16, 1)
    )

def gen_noise(batch_size, noise_dim, device):
    return torch.randn(batch_size, noise_dim, device=device)

def train_step(real_images, generator, discriminator, gen_opt, disc_opt, loss_func, device):
    batch_size = real_images.size(0)
    noise_dim = 100

    # Train Discriminator
    disc_opt.zero_grad()
    noise = gen_noise(batch_size, noise_dim, device)
    fake_images = generator(noise)

    real_preds = discriminator(real_images)
    fake_preds = discriminator(fake_images.detach())

    real_loss = loss_func(real_preds, torch.ones_like(real_preds))
    fake_loss = loss_func(fake_preds, torch.zeros_like(fake_preds))
    disc_loss = (real_loss + fake_loss) / 2

    disc_loss.backward()
    disc_opt.step()

    # Train Generator
    gen_opt.zero_grad()
    noise = gen_noise(batch_size, noise_dim, device)
    fake_images = generator(noise)
    fake_preds = discriminator(fake_images)

    gen_loss = loss_func(fake_preds, torch.ones_like(fake_preds))

    gen_loss.backward()
    gen_opt.step()

    return gen_loss.item(), disc_loss.item()

def main():
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

    # Load Anime Face dataset
    transform = transforms.Compose([
        transforms.Resize((64, 64)),
        transforms.ToTensor(),
        transforms.Normalize((0.5, 0.5, 0.5), (0.5, 0.5, 0.5))  # Normalize to [-1, 1]
    ])

    dataset_path = "/root/.cache/kagglehub/datasets/splcher/animefacedataset/versions/3"
    train_dataset = datasets.ImageFolder(root=dataset_path, transform=transform)
    train_loader = DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)

    generator = make_generator_model().to(device)
    discriminator = make_discriminator_model().to(device)

    gen_opt = optim.Adam(generator.parameters(), lr=1e-4)
    disc_opt = optim.Adam(discriminator.parameters(), lr=1e-4)

    loss_func = nn.BCEWithLogitsLoss()

    # Training parameters
    noise_dim = 100
    num_examples_to_generate = 16
    seed = gen_noise(num_examples_to_generate, noise_dim, device)

    # Training loop
    start_time = time.time()
    for epoch in range(EPOCHS):
        epoch_start_time = time.time()
        gen_loss_avg = 0
        disc_loss_avg = 0
        num_batches = 0

        for real_images, _ in train_loader:
            real_images = real_images.to(device)
            gen_loss, disc_loss = train_step(real_images, generator, discriminator, gen_opt, disc_opt, loss_func, device)

            gen_loss_avg += gen_loss
            disc_loss_avg += disc_loss
            num_batches += 1

        gen_loss_avg /= num_batches
        disc_loss_avg /= num_batches

        epoch_elapsed_time = time.time() - epoch_start_time
        print(f"Epoch {epoch+1}/{EPOCHS} - Generator Loss: {gen_loss_avg:.4f}, Discriminator Loss: {disc_loss_avg:.4f}, Time: {epoch_elapsed_time:.2f}s")

        # Generate and save images after every 10 epochs
        if (epoch + 1) % 10 == 0:
            with torch.no_grad():
                generated_images = generator(seed).cpu()
                generated_images = (generated_images * 0.5 + 0.5).clamp(0, 1).numpy()

            fig, axes = plt.subplots(4, 4, figsize=(4, 4))
            for i, ax in enumerate(axes.flat):
                ax.imshow(np.transpose(generated_images[i], (1, 2, 0)))
                ax.axis("off")
            plt.savefig(f"generated_images_epoch_{epoch+1}.png")
            plt.show()

    total_elapsed_time = time.time() - start_time
    print(f"Training completed in {total_elapsed_time:.2f}s")

if __name__ == "__main__":
    BATCH_SIZE = 128
    EPOCHS = 50
    main()
```
